# 场景案例显神通

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
```

```python
drug=pd.read_csv('./Drugs.csv')
drug.rename(columns={'YYYY':'Year'},inplace=True)
```

```python
drug.head()
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,1,1)
sns.countplot('Year',data=drug,palette='RdYlGn_r',edgecolor=sns.color_palette('dark',7))
plt.subplot(2,1,2)
sns.barplot(drug['SubstanceName'].value_counts()[:15].index,drug['SubstanceName'].value_counts()[:15].values,palette='inferno')
```
