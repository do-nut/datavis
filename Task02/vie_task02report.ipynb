```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig,ax=plt.subplots(2,1,figsize=(8,6),sharex=True,facecolor='#FFFAF0',frameon=True,constrained_layout=True)
plt.subplots_adjust(hspace=0.2,wspace=1,right=1.5,left=0.7)
x=np.linspace(0,15,100)
ax[1].plot(x,np.sin(x))
ax[1].patch.set_facecolor('#98F5FF')
ax[1].patch.set_alpha(0.8)
ax[1].grid(color='w')
ax[1].spines['right'].set_color(None)
ax[1].spines['top'].set_color(None)
ax[1].tick_params(which='both',length=0)
ax[0].minorticks_on()
ax[0].tick_params(which='both',top='on',right='on',labelright='on',rotation=45,labelcolor='r',color='b')
```

```python
fig=plt.figure(figsize=(12,10))
x1 = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
x2 = np.arange(0, 2 * np.pi, 0.02)

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2, polar=True)
ax3 = fig.add_subplot(2, 2, 3, polar=True)
ax1.plot(x1,np.sin(x1))
ax2.plot(x2,np.sin(x2))
ax2.set_yticks([-1,-0.5,0,0.5,1],minor=False)
ax2.set_ylim(-1,1)
ax3.bar(x1,x1,color='yellow', edgecolor='red',alpha=0.5,width=0.5)
```

## 思考题
****
1. primitives和container的区别和联系
+ container是一个对象，primitives则是属性，有点类似于cpp中的类和成员
+ 联系自然也就是primitives构成了完整的container
2. 四个容器的联系和区别是什么？他们分别控制一张图的哪些要素？
+ figure、axes、axis、tick四大容器共同构成了一个完整的图像展示
+ figure可以看做一个大的展示墙，而axes则是一幅幅裱起来的画，axis和tick则代表着这幅画框的属性，分别控制着呈现结果的不同区域
+ figure控制整体的画布，axes控制小一点的画，也可以说子图，axis则是子图的边框属性，tick也属于边缘属性，和axis分别控制着不同的边缘属性

```python
x1=np.linspace(0,15,100)
x2=np.linspace(0,15,20)
x3=np.random.randint(0,100,100)
fig=plt.figure(figsize=(12,10))
#‘aitoff’, ‘hammer’, ‘lambert’, ‘mollweide’, ‘polar’, ‘rectilinear’
ax1 = fig.add_subplot(2, 3, 1)
ax2 = fig.add_subplot(2, 3, 2)
ax3 = fig.add_subplot(2, 3, 3, projection='polar')
ax4 = fig.add_subplot(2, 3, 4, projection='hammer')
ax5 = fig.add_subplot(2, 3, 5, projection='lambert')
ax6 = fig.add_subplot(2, 3, 6, projection='rectilinear')
ax1.plot(x1,np.sin(x1))
ax2.scatter(x2,x2)
ax4.set_xticks(np.arange(-2,3,1))
ax5.set_xticks(np.arange(-2,3,1))
ax6.hist(x3)
plt.suptitle('collection',fontsize=24)
```
